```python
!pip install mysql-connector-python
import pandas as pd
import numpy as np
import mysql.connector

from google.colab import drive
drive.mount('/content/drive')
```

```python
users = pd.read_csv('/content/drive/MyDrive/DataScience/users.csv',sep='\t').drop('Unnamed: 0',axis='columns')
users
```

```python
ages = pd.read_csv('/content/drive/MyDrive/DataScience/age_ranges.csv',sep='\t').drop('Unnamed: 0',axis='columns')
occupations = pd.read_csv('/content/drive/MyDrive/DataScience/occupations.csv',sep='\t').drop('Unnamed: 0',axis='columns')

ageDict = ages.to_dict()['ageRange']
occDict = occupations.to_dict()['description']

users['age'] = users['age'].replace(ageDict)
users['occupation'] = users['occupation'].replace(occDict)

users
```

```python
users.to_csv('/content/drive/MyDrive/DataScience/usersMerged.csv',sep='\t',encoding='utf-8')
```
