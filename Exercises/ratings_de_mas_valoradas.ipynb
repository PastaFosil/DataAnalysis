```python
ratings = pd.read_csv('/content/drive/MyDrive/DataScience/ratings.csv',sep='\t').drop('Unnamed: 0',axis='columns')
movies = pd.read_csv('/content/drive/MyDrive/DataScience/movies.csv',sep='\t').drop('Unnamed: 0',axis='columns')
ratings
```

```python
count = ratings.groupby(['movieID']).size()

ratings_filtrados_con_nombre = pd.DataFrame(count[count>2500],columns=['ratingCount']).reset_index('movieID')
ratings_filtrados_con_nombre = ratings_filtrados_con_nombre.merge(movies[['movieID','title']])

ratings_filtrados_con_nombre['movieID']
```

```python
def is_in_list(id):
  count = ratings.groupby(['movieID']).size()

  ratings_filtrados_con_nombre = pd.DataFrame(count[count>2500],columns=['ratingCount']).reset_index('movieID')
  ratings_filtrados_con_nombre = ratings_filtrados_con_nombre.merge(movies[['movieID','title']])
  li = list(ratings_filtrados_con_nombre['movieID'])

  return id in li
```

```python
ratings.head(50)[ratings.head(50)['movieID'].apply(is_in_list)]
```
