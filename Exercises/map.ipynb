```python
import numpy as np
import pandas as pd
```

```python
def booleanNum(boo):
  if boo==True:
    return 1
  else:
    return 0
```

```python
df_reto_3 = pd.read_csv('/content/drive/MyDrive/near_earth_objects-jan_feb_1995-reto_2.csv',sep=',')

#df_reto_3['is_potentially_hazardous_asteroid'].map(booleanNum)

boo = {True: 1, False: 0}
df_reto_3['is_potentially_hazardous_asteroid'] = df_reto_3['is_potentially_hazardous_asteroid'].replace(boo)

```

```python
def kmh_to_kmmin(vel):
  return vel/60
```

```python
df_reto_3['relative_velocity.kilometers_per_minute'] = df_reto_3['relative_velocity.kilometers_per_hour'].map(kmh_to_kmmin)
```

```python
df_reto_3.to_csv('near_earth_objects-jan_feb_1995-reto_3.csv',sep=',',encoding='utf-8')
```
